# Amazon Forecast

In this lab, we will try to use [Amaozn Forecast](https://docs.aws.amazon.com/forecast/latest/dg/what-is-forecast.html), a fully managed service that uses machine learning to deliver highly accurate forecasts.

Amzon Forecast offer 5 algorithm now, they are ARIMA, DeepAR+, ETS, NPTS and Prophet, you can use one of them or enable **Auto ML** option to let Amzon Forecast choose algorithm for you. In this lab, we will use Prophet for example, you can use other algorithm as you want.

## Prerequest

Make sure you have a proper permission to use Amazon Forecast, can create IAM role for Amzon Forecast. You can follow follow link to find out how to create IAM role.

https://docs.aws.amazon.com/forecast/latest/dg/aws-forecast-iam-roles.html#aws-forecast-create-iam-role-with-console

### Data download

We have processed the data for this lab, you can download data from below link and upload to your S3 bucket for Amazon Forecast, make sure the IAM role you created before have the permission to read objects in this S3 bucket.

**Target time series files**: https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/data/forecast-timeseries-data.csv

Data looks like:

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/data.png)

The date range from 2019-07-10 to 2019-09-23, store id from store0 to store1603. We would predict from 2019-09-24 to 2019-09-30.

## Create Dataset Group and Import Dataset

1. Go to Amzon Forecast console, currently Amzon Forecast is available in **us-east-1, us-east-2, us-west-2, ap-southeast-1, ap-northeast-1, eu-west-1**

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/console.png)

Choose **Create dataset group**

Under Step1, choose **Metrics** in Forecasting domain:

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step1.png)

Step2, keep **day** in Your data entires have a time interval, and modify Data schema as below.

<img src="https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step2.png" style="zoom:67%;" />

 Schema:

``` json
{
	"Attributes": [
		{
			"AttributeName": "timestamp",
			"AttributeType": "timestamp"
		},
		{
			"AttributeName": "metric_name",
			"AttributeType": "string"
		},
		{
			"AttributeName": "metric_value",
			"AttributeType": "float"
		}
	]
}
```

Next, modify Timestamp format to **yyyy-MM-dd** , choose IAM role you created before and specify the data location in your S3 bucket (use S3 path format).

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step3.png)

If everything is OK, then **Start Import**, you would see below.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step4.png)

You can switch to **Datasets** console, to check the import progress.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step5.png)

This could take 5 mins. 

When it finished, the Status and Latest import status would be **Active**.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step6.png)

 

## Create predictor

Go to Dashboard, click orange button **Start**:

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step7.png)

Value for Predictors:

Forecast horizon: 7

Algorithm selection: Manual

Algorithm: Prophet (In this lab, we use Prophet for sample)

Keep other settings to default, then start.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step8.png)

Go to Predictors, you can check the training progress.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step9.png)



This would takes about 30 mins.

## Create Forecast

When the Training status turned to Active, we can create a Forecast.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step10.png)

Enter a Forcast name, keep others to default.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step11.png)

When the Status turned to Active, we can query forecast. The creation takes about 30 mins.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step12.png)



## Query Forecast results

Go to **Forecast lookup**, under Forecast lookup, set Start date **2019/09/23, End date **2019/09/30**, Value could use from **store0** to **store1603**, here we could choose **store24, store31, store36, store87**, the same sample sites as the lab1~3,  click **Get Forecast**, you could get metrics of P10, P50, P90.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step13.png)



![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step14.png)

You can export complete forecast result to S3 bucket by **Create forecast export**, after export job finishes, you can get the report from S3.

![](https://tko-2020-ts-workshop.s3-us-west-2.amazonaws.com/forecast/image/step15.png)

## Clean Up Resources

To avoid unnecessary cost, remember clean up Amzon Forecast resources:

Delete **Forecasts**, then delete **Predictiors**, then delete **Datasets** and **Dateset group**.
